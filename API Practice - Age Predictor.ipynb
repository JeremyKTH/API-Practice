API Practice 02: The Age Predictor

```python
import requests

# API endpoint for age prediction
url_ageify = "https://api.agify.io/"

response1 = requests.get(url_ageify, params={'name': 'Jeremy'})

print("Response 1:", response1.json())
```

```python
print("His name is: ", response1.json()["name"] )
print("His predicted age is: ", response1.json()["age"] )
```
