API practice 01: Dog API

```python
import requests
```

```python
# URL to get a random dog image
url_RandomDogImage = "https://dog.ceo/api/breeds/image/random"

# Make a GET request to the API to retrieve a random dog image
response = requests.get(url_RandomDogImage)

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON response
    data = response.json()
    
    print(data["message"])
```

```python
# URL to get a list of all dog breeds
url_AllBreeds = "https://dog.ceo/api/breeds/list/all"

# Make a GET request to the API to retrieve a list of all dog breeds
response = requests.get(url_AllBreeds)

print(response.json())
```
