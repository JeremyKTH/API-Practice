### Popular Movies (GET)

```python
# Popular Movies Endpoint
url_popularMovies = "https://api.themoviedb.org/3/movie/popular"

headers = {
    "accept": "application/json",
    "authorization": f"Bearer {token}"
}

# Endpoint: English, Page 1
params = {
    "language": "en-US",
    "page": 1,
    "region": "Sweden"
}
```

#### Adding params to the request
Two ways:
1. f"url_popularMovies"{?langauge=en-US&page=1&region=Sweden}
2. Create dictionary params, and add into get()

```python
# Add the params to the request
response = requests.get(url_popularMovies, headers= headers, params=params)
data = response.json()

print(data["total_results"])
```

```python
# Get the first movie from the results

first_movie_data = data["results"][0]

# original title
print(f"Original Title: {first_movie_data['original_title']}")

# overview
print(f"Overview: {first_movie_data['overview']}")

# release date
print(f"Release Date: {first_movie_data['release_date']}")
```
