## Practice API POST and PUT method with Sandbox httpbin

#### POST (Create): You send data to a general category to create something brand new.

```python
import requests

# Sandbox endpoint
url = "https://httpbin.org/post"

# Payload (data to send)
payload = {
    "movie_title": "Inception",
    "movie_id": 27205,
    "is_favorite": True
}

# Make the POST request
response = requests.post(url, json=payload)

server_respose = response.json()
print(server_respose)
```

#### PUT VS PATCH (Update): 

PUT: You send data to overwrite the entire object.

PATCH: You send data to update one piece of information without touching the rest of the object.
